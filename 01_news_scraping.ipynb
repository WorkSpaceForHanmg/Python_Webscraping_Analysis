```python
import requests
import bs4 
# BeautifulSoup 클래스 import
from bs4 import BeautifulSoup


#불러다가 쓸꺼니까 import 한다.
```

#1-1

```python


   
req_param = {
    'sid': 'economy'         #딕셔너리 
                        #sid 번호에 따라 사이트가 바뀌기에
}
# 
url = 'https://news.daum.net/{sid}'.format(**req_param) #딕셔너리 넣을땐 앞에 **추가
print(url)


# 요청 헤더 설정 : 브라우저 정보 (사람처럼 보이게 하려고 설정하는 것임)
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'
}

# requests 의 get() 함수 호출하기 
res = requests.get(url, headers=req_header)
print(res.status_code)
res.encoding = 'utf-8'
# 응답(response)이 OK 이면
if res.ok:
    # 응답 (response)에서 text 추출 - source 보기
    html = res.text
    # print(html)
    # BeautifulSoup 객체 생성  
    soup = BeautifulSoup(html, 'html.parser')
    #print(soup)
    
    # CSS 선택자
    a_tag_list = soup.select("ul.list_newsheadline2 li a[href*='daum.net/v']")
    print(type(a_tag_list),(len(soup.select("ul.list_newsheadline2 li a[href*='daum.net/v']")))) 
    
    
    
    for a_tag in a_tag_list:
        #print(type(a_tag))
        title = a_tag.text.strip() #strip() 공백을 제거하는 함수
        link = a_tag['href']   #herf 속성의 값을 가져온ㄷ나
        print(link)
        print(title) 
        
    # <a> 태그 리스트 순회하기    
else:
    # 응답(response)이 Error 이면 status code 출력    
    print(f'에러 코드 = {res.status_code}')
 
```

```python
section_dict = {
    "기후/환경": "climate",
    "사회": "society",
    "경제": "economy",
    "정치": "politics",
    "국재": "world",
    "문화": "culture",
    "생활": "life",
    "IT/과학": "tech",
    "인물": "people",
}
def print_news(sid):  # print_new(103)
    section = section_dict[sid]
    req_param = {"sid": section}


    url = "https://news.daum.net/{sid}".format(**req_param)  # 딕셔너리 넣을땐 앞에 **추가
    print(f'=====>{url} {sid}뉴스<=====')

    # url2 = 'https://news.daum.net/economy'

    #요청 헤더 설정 : 브라우저 정보 (사람처럼 보이게 하려고 설정하는 것임)
    req_header = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36"
    }

    # requests 의 get() 함수 호출하기
    res = requests.get(url, headers=req_header)
    res.encoding = 'utf-8'
    # 응답(response)이 OK 이면
    if res.ok:
            # 응답 (response)에서 text 추출 - source 보기
            html = res.text
            # print(html)
            # BeautifulSoup 객체 생성  
            soup = BeautifulSoup(html, 'html.parser')
            #print(soup)
            
            # CSS 선택자
            
            #print(len(soup.select("ul.list_newsheadline2 li a[href*='daum.net/v']")))
            
            a_tag_list = soup.select("ul.list_newsheadline2 li a[href*='daum.net/v']")
            #print(type(a_tag_list)) # [<a href ="http://">뉴스제목</a>] ResultSet객체 , <a></a> : Tag 객체
            #print(a_tag_list) #[<a></a>,<a></a>]
            for a_tag in a_tag_list:
                #print(type(a_tag))
                title = a_tag.text.strip() #strip() 공백을 제거하는 함수
                link = a_tag['href']   #herf 속성의 값을 가져온ㄷ나
                print(link)
                print(title) 
            print()
            
    else:
        # 응답(response)이 Error 이면 status code 출력 
        print(f'에러 코드 = {res.status_code}')
        
        
print_news('경제')
print_news('기후/환경')
```
