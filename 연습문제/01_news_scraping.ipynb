#1-1

```python


# 요청 파라미터
req_param = {
    'sid': 'economy'         # 딕셔너리
}

# URL 생성
url = 'https://news.daum.net/{sid}'.format(**req_param)
print(url)

# 요청 헤더 설정
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'
}

# GET 요청
res = requests.get(url, headers=req_header)
print(res.status_code)

res.encoding = 'utf-8'

# 응답 OK인 경우
if res.ok:
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')

    # li 태그 리스트 추출
    # [li]
    li_tag_list = soup.select("ul.list_newsheadline2 li")
    print(type(li_tag_list), len(li_tag_list))

    for li_tag in li_tag_list:
            a_tag = li_tag.find("a")
            title = li_tag.select_one("div.cont_thumb  strong.tit_txt").text.strip()  # strip() 공백을 제거하는 함수
            link = a_tag["href"]  # href 속성의 값을 가져온ㄷ나
            print(link)
            print(title)
        

# 응답 실패한 경우
else:
    print(f'에러 코드 = {res.status_code}')
```

```python
section_dict = {
    "기후/환경": "climate",
    "사회": "society",
    "경제": "economy",
    "정치": "politics",
    "국제": "world",
    "문화": "culture",
    "생활": "life",
    "IT/과학": "tech",
    "인물": "people",
}


def print_news(sid): 
    section = section_dict[sid]
    req_param = {"sid": section}

    url = "https://news.daum.net/{sid}".format(
        **req_param
    )  # 딕셔너리 넣을땐 앞에 **추가
    print(f"=====>{url} {sid}뉴스<=====")

    # url2 = 'https://news.daum.net/economy'

    # 요청 헤더 설정 : 브라우저 정보 (사람처럼 보이게 하려고 설정하는 것임)
    req_header = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36"
    }

    # requests 의 get() 함수 호출하기
    res = requests.get(url, headers=req_header)
    res.encoding = "utf-8"
    # 응답(response)이 OK 이면
    if res.ok:
        # 응답 (response)에서 text 추출 - source 보기
        html = res.text

        # BeautifulSoup 객체 생성
        soup = BeautifulSoup(html, "html.parser")

        li_tag_list = soup.select("ul.list_newsheadline2 li")

        for li_tag in li_tag_list:
            a_tag = li_tag.find("a")
            title = li_tag.select_one("div.cont_thumb  strong.tit_txt").text.strip()  # strip() 공백을 제거하는 함수
            link = a_tag["href"]  # href 속성의 값을 가져온ㄷ나
            print(link)
            print(title)
        print()

    else:
        # 응답(response)이 Error 이면 status code 출력
        print(f"에러 코드 = {res.status_code}")


print_news("경제")
print_news("기후/환경")
```

```python
import requests
from bs4 import BeautifulSoup

def print_news(section_name):    
    #section = section_dict[section_name]
    section = section_dict.get(section_name)
    if section is None:
        print('요청하신 Key가 없습니다.')
    else:    
        # 요청 Parameter
        req_param = {
            'section': section
        }
        url = 'https://news.daum.net/{section}'.format(**req_param)
        
        print(f'======> {url} {section_name} 뉴스 <======')
        
        # 요청 헤더 설정 : 브라우저 정보
        req_header = {
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'
        }

        res = requests.get(url, headers=req_header)   

        if res.ok:
            res.encoding = 'utf-8' 
            html = res.text
            soup = BeautifulSoup(html, 'html.parser')
            
            # 기사 제목과 링크 추출
            li_tag_list = soup.select('ul.list_newsheadline2 li')

            #print(type(li_tag_list), len(li_tag_list))

            for li_tag in li_tag_list:        
                a_tag = li_tag.find('a')
                print(a_tag['href'])
                
                strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
                title = strong_tag.text.strip()
                print(title)

        else:
            print(f'에러코드 = {res.status_code}')
            
print_news('기후/환경')
```
